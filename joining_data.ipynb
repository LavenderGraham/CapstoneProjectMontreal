```python
import geohash
```

```python
# Define a function that takes latitude and longitude and returns a six-character geohash
def generate_geohash6(lat, lon):
    return geohash.encode(lat, lon, precision=6)
```

```python
# Add precision 6 geohashes
montreal_transport_stations['Geohash_prec6'] = montreal_transport_stations.apply(lambda row: generate_geohash6(row['Latitude'], row['Longitude']), axis=1)
```

```python
montreal_transport_stations
```

```python
montreal_all_rail_stations['Geohash_prec6'] = montreal_all_rail_stations.apply(lambda row: generate_geohash6(row['Latitude'], row['Longitude']), axis=1)
```

```python
montreal_all_rail_stations
```

```python
bars_and_restaurants_complete['Geohash_prec6'] = bars_and_restaurants_complete.apply(lambda row: generate_geohash6(row['Latitude'], row['Longitude']), axis=1)
```
